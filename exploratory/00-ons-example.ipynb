```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# define data paths
air_emmisions_dataset = '../data/ons-air-emissions-all-greenhouse-gases.csv'
air_travel_dataset = '../data/ons-air-travel.csv'
```

```python
# load datasets
air_emmisions_df = pd.read_csv(air_emmisions_dataset, skiprows=8, names=['Year', 'CO2-thousand-tonnes'])
air_travel_df = pd.read_csv(air_travel_dataset, skiprows=8, names=['Year', 'Passenger-Air-Transport-Services'])
# keep only the year summaries & set to year
air_travel_df = air_travel_df.iloc[:13]
air_travel_df['Year'] = air_travel_df['Year'].astype(int)
# CO2 to million tonnes
air_emmisions_df['CO2-thousand-tonnes'] = air_emmisions_df['CO2-thousand-tonnes'].astype(float) / 1000
```

```python
# merge datasets
air_df = air_emmisions_df.merge(air_travel_df, on='Year')
```

```python

# plot datasets
sns.set(style="darkgrid")
plt.gcf().set_size_inches(10, 6)
sns.lineplot(x="Year", y="CO2-thousand-tonnes", data=air_df)
sns.lineplot(x="Year", y="Passenger-Air-Transport-Services", data=air_df)
plt.legend(labels=['CO2 million tonnes', 'Passenger air transport services'])
plt.ylabel('')
```
