```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# dataset file
mean_temperature_dataset = '../data/met-office-England_N-mean-air-temperature.txt'
```

```python
# load dataset
mean_temp_df = pd.read_csv(mean_temperature_dataset, skiprows=5, sep='\s+', engine='python')
```

```python
# plot
sns.set_style('whitegrid')
plt.gcf().set_size_inches(10, 6)
sns.lineplot(x="year", y="ann", data=mean_temp_df)
# fit a trend line
plt.plot(mean_temp_df.year, mean_temp_df.ann.rolling(window=10).mean(), color='red', linewidth=2)
plt.ylabel('Temperature (°C)')
plt.xlabel('Year')
plt.title('Mean annual temperature: North of England')
```
